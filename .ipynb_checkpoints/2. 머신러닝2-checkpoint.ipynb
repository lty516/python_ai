```python
# 기본 모듈 임포트
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns 

import matplotlib.font_manager as fm 
font_name = fm.FontProperties(fname="C:\\Windows\\Fonts\\malgun.ttf").get_name()
plt.rc("font",family=font_name)

mpl.rcParams["axes.unicode_minus"] = False

#------------------------------------------------------------------------------

from sklearn.model_selection import train_test_split

#pip install mglearn
import mglearn      # 머신러닝 학습을 위한 모듈
```

```python
#샘플 데이터 불러오기
mglearn.plots.plot_linear_regression_wave()
```

```python
from sklearn.linear_model import LinearRegression

# 데이터를 변수에 담기
X,y = mglearn.datasets.make_wave(n_samples=60)

# 훈련 데이터, 테스트데이터 나누기 
X_train,X_test,y_train,y_test = train_test_split(X, y, random_state = 42)

# 학습 
lr = LinearRegression().fit(X_train,y_train)
```

```python
# 결과 확인 coef_: 기울기 / intercept_: 절편 / score: 결과확인
print(lr.coef_,lr.intercept_)

# 성능 측정 
print(lr.score(X_train,y_train))
print(lr.score(X_test,y_test))
```

```python
# 가설
# 주제 : 전기 생산량과 소비량  (상관 관계 파악)
# 귀무 가설(연구가치가 없다.) : 전기 생산량과 소비량은 
```

```python
# 데이터
# 월별 전기 사용량(백만kwh)
y = [2.48, 2.27, 2.47, 2.77, 2.98, 3.05, 3.18, 3.46, 3.03, 3.25, 2.67, 2.53]

# 월별 전기 생산금액(억원)
X = [3.52, 2.58, 3.31, 4.07, 4.62, 3.98, 4.29, 4.83, 3.71, 4.61, 3.90, 3.20]
```

```python
# 데이터마이닝 전문 모듈   # 가설을 결정할 때 pvalue 확인 5% 이하면 상관관계가 있음(대립 가설)  
# >> 선형회귀할 가치가 있다.
from scipy import stats

result = stats.linregress(X,y)
result
```

```python
# lr = LinearRegression().fit(X,y)     >    X를 2차원으로 변환시켜야 한다.
num = np.array(X)
num=num.reshape(-1,1)
# 학습
lr = LinearRegression().fit(num,y) 
```

```python
# 결과확인
lr.coef_,lr.intercept_
```

```python
plt.scatter(X,y)
# 회귀선 그리기  wx + b
plt.plot(np.array(X),lr.coef_*np.array(X)+lr.intercept_,c='red')
```

```python
# 생산량이 5일때 사용량은 얼마일까?
5 * lr.coef_+lr.intercept_
```

## 선형회귀 모델의 종류
<hr/>

###  릿지(Ridge) 회귀 모델
   
   + L2 규제(Regularization)
   + MSE항에 Penalty항 추가 
   + 알파를 크게하면 패널티의 효과가 커지고(가중치 감소), 작게하면 그 반대가 된다.
   
###  라쏘(Lasso) 회귀 모델

   + L1 규제(Regularization)
   + 알파를 크게하면 패널티의 효과가 커지고(가중치 감소), 작게하면 그 반대가 된다.
   + 특성 선택이 자동으로 이루어진다.
 
###  엘라스틱넷(ElasticNet) 회귀 모델

   + Ridge + Lasso

### 릿지회귀

```python
from sklearn.linear_model import Ridge

ridge = Ridge().fit(X_train,y_train) # alpha = 1.0 < default
# 성능 측정 
print("훈련 점수 : ",ridge.score(X_train,y_train))
print("테스트 점수 : ",ridge.score(X_test,y_test))  
```

```python
# 알파값 증가
ridge10 = Ridge(alpha = 10).fit(X_train,y_train)
# 성능 측정 
print("훈련 점수 : ",ridge10.score(X_train,y_train))
print("테스트 점수 : ",ridge10.score(X_test,y_test))  
```

```python
# 알파값 감소
ridge01 = Ridge(alpha = 0.1).fit(X_train,y_train)
# 성능 측정 
print("훈련 점수 : ",ridge01.score(X_train,y_train))
print("테스트 점수 : ",ridge01.score(X_test,y_test))  
```

```python
# 시각화
plt.plot(ridge10.coef_, '^', label="Ridge alpha=10")
plt.plot(ridge.coef_, 's', label="Ridge alpha=1")
plt.plot(ridge01.coef_, 'v', label="Ridge alpha=0.1")

plt.plot(lr.coef_, 'o', label="LinearRegression")
plt.xlabel("계수 목록")
plt.ylabel("계수 크기")
plt.hlines(0, 0, len(lr.coef_))
plt.ylim(-25, 25)
plt.legend()
```

```python
# 알파값을 1로 고정시킨 상태에서 데이터의 크기 조절
mglearn.plots.plot_ridge_n_samples()
```

### 라쏘 회귀

```python
from sklearn.linear_model import Lasso

lasso = Lasso().fit(X_train,y_train) # alpha = 1.0 < default
# 성능 측정 
print("훈련 점수 : ",lasso.score(X_train,y_train))
print("테스트 점수 : ",lasso.score(X_test,y_test))  
print("사용한 특성의 수 : ",np.sum(lasso.coef_ != 0))
```

```python
# 알파값 감소
lasso001 = Lasso(alpha = 0.01).fit(X_train,y_train)
# 성능 측정 
print("훈련 점수 : ",lasso001.score(X_train,y_train))
print("테스트 점수 : ",lasso001.score(X_test,y_test))  
print("사용한 특성의 수 : ",np.sum(lasso001.coef_ != 0))
```

```python
# 알파값 감소
lasso00001 = Lasso(alpha = 0.0001).fit(X_train,y_train)
# 성능 측정 
print("훈련 점수 : ",lasso00001.score(X_train,y_train))
print("테스트 점수 : ",lasso00001.score(X_test,y_test))  
print("사용한 특성의 수 : ",np.sum(lasso00001.coef_ != 0))
```

```python
# 시그모이드 함수 : 0과 1사이에 값을 출력  p.193

def sigmoid(x):
    result = []
    for i in x:
        result.append(1/(1 + np.exp(-i)))   # np.exp :지수 함수 
    return result

# 데이터
x = np.linspace(-8,8,100)
sig = sigmoid(x)
plt.plot(x,sig)
plt.plot(x,np.tanh(x)) # 범위를 증가 -1 ~ 1
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC 
```

```python
X, y = mglearn.datasets.make_forge()
fig, axes = plt.subplots(1, 2, figsize=(10, 3))

for model, ax in zip([LinearSVC(), LogisticRegression()], axes):
    clf = model.fit(X, y)
    
    mglearn.plots.plot_2d_separator(clf, X, fill=False, eps=0.5, ax=ax, alpha=.7)
    mglearn.discrete_scatter(X[:, 0], X[:, 1], y, ax=ax)
    
    ax.set_title("{}".format(clf.__class__.__name__))
    ax.set_xlabel("특성 0")
    ax.set_ylabel("특성 1")
```

```python
# c값을 조절을 하여 일반화 / 복잡하게 튜닝된다.
mglearn.plots.plot_linear_svc_regularization()

# 규제가 증가 : 내가 많이 간섭을해서 복잡도를 낮춘다
# 규제가 감소 : 내가 덜 간섭해서 복잡도가 높아진다.
# c값이 낮아지면 규제가 증가   
# c값이 높아지면 규제가 감소
# 규제 : 과대적합을 막기위해 > 모델이 복잡해진다
```

```python
# 실전예제
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
# 훈련/테스트데이터 분리
X_train,X_test,y_train,y_test = train_test_split(cancer.data, cancer.target, random_state = 42, stratify = cancer.target)
```

```python
logreg = LogisticRegression().fit(X_train,y_train)   # C = 1 , penalty ='l2'   < default
print("훈련 점수 : ",logreg.score(X_train,y_train))
print("테스트 점수 : ",logreg.score(X_test,y_test))

# 훈련과 테스트의 결과가 차이가 거의없다면 과소적합을 의심해 볼 수 있다.
```

```python
logreg100 = LogisticRegression(C=100, penalty='l2').fit(X_train,y_train)  
print("훈련 점수 : ",logreg100.score(X_train,y_train))
print("테스트 점수 : ",logreg100.score(X_test,y_test))
```

```python
logreg001 = LogisticRegression(C=0.01, penalty ='l2').fit(X_train,y_train)
print("훈련 점수 : ",logreg001.score(X_train,y_train))
print("테스트 점수 : ",logreg001.score(X_test,y_test))
```

```python
# L1 > 모델을 단순화 (극단적)
# penalty= 'l1'사용시 solver='liblinear'를 함께 사용
logreg_l1 = LogisticRegression(C=1, penalty ='l1', solver='liblinear').fit(X_train,y_train)
print("훈련 점수 : ",logreg_l1.score(X_train,y_train))
print("테스트 점수 : ",logreg_l1.score(X_test,y_test))
```

```python
logreg100_l1 = LogisticRegression(C=100, penalty ='l1', solver='liblinear').fit(X_train,y_train)
print("훈련 점수 : ",logreg100_l1.score(X_train,y_train))
print("테스트 점수 : ",logreg100_l1.score(X_test,y_test))
```

```python
logreg001_l1 = LogisticRegression(C=0.01, penalty ='l1', solver='liblinear').fit(X_train,y_train)
print("훈련 점수 : ",logreg001_l1.score(X_train,y_train))
print("테스트 점수 : ",logreg001_l1.score(X_test,y_test))
```

```python
df = pd.read_csv("data/train.csv")
df.head()
```

```python
# 전처리  컴퓨터 언어로 변환(숫자로 변환)
# 전처리가 필요없는 것들 
cols_to_keep = ["Survived","Age","Fare"]
```

```python
# one-hot encoding : 하나의 값을 강조 >> pd.get_dummies
dummy_class = pd.get_dummies(df["Pclass"],prefix="Pclass")   # prefix > 접두사 
dummy_sex = pd.get_dummies(df["Sex"],prefix="Sex")

display(dummy_class.head(),dummy_sex.head())
```

```python
# 전처리한 데이터 합치기
df1 = df[cols_to_keep].join(dummy_class.loc[:,"Pclass1":])
df1.head()
```

```python
# 성별은 하나의 데이터만 있어도 구별 가능하기 때문에 하나의 데이터만 불러옴 
df1 = df1.join(dummy_sex.loc[:,"Sex_male":])
```

```python
train_cols = df1.columns[1:]
df1[train_cols]
```

```python
# 결측치 확인 
np.isnan(df1[train_cols])
```

```python
# 결측치 삭제를 한다면.. dropna
# df1[train_cols] = df1[train_cols].dropna(axis=0)
# df1["Survived"] = df1["Survived"].dropna(axis=0)

# 결측치 0으로 대체  fillna()
df1[train_cols] = df1[train_cols].fillna(0)
df1["Survived"] = df1["Survived"].fillna(0)
```

```python
# 훈련 + 서머리 statsmodels.api
import statsmodels.api as sm
# 훈련 시작   sm.Logit(레이블 , 훈련데이터).fit()
model = sm.Logit(df1["Survived"],df1[train_cols]).fit()
# 기술 통계족 분석용도 / 실제 머신러닝으로써 훈련시키는 것은 아님
```
